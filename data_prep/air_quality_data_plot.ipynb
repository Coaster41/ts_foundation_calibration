```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_air = pd.read_csv('../data/air+quality/AirQuality.csv', decimal=',', na_values=-200)
df_air
```

```python
custom_format = '%d/%m/%Y%H.%M.%S'
df_air.insert(0,'ds', pd.to_datetime(df_air['Date'] + df_air["Time"], format=custom_format))
df_air
```

```python
column_dict = []
dfs = []
for i, column in enumerate(df_air.columns[3:]):
    column_dict.append([column, i])
    df_temp = df_air[['ds', column]].copy()
    df_temp = df_temp.rename(columns={column: 'y'})
    df_temp['unique_id'] = i
    dfs.append(df_temp)
    # display(df_temp)
df = pd.concat(dfs)
df = df.reset_index(drop=True)
df
```

```python
def plot_df(df, id, timestamp, graph_before, steps_ahead, plt_tmp=plt):
    true_y = df.loc[(df["unique_id"]==id) & (df["ds"]>=timestamp-pd.Timedelta(graph_before, unit='h')) & (df["ds"]<timestamp+pd.Timedelta(max(steps_ahead,24), unit='h')), "y"].to_numpy().flatten()
    plt_tmp.plot(df.loc[(df["unique_id"]==id) & (df["ds"]>=timestamp-pd.Timedelta(graph_before, unit='h')) & (df["ds"]<timestamp+pd.Timedelta(max(steps_ahead,24), unit='h')), "ds"].to_numpy().flatten(), true_y, label=f"True", linewidth=1, color='r')

plt.figure(figsize=(24,6))
id = 3
timestamp = pd.Timestamp("2004-08-12 00:00:00")
graph_before = 24*7*3
steps_ahead = 24
plot_df(df, id, timestamp, graph_before, steps_ahead)
plt.ylabel(column_dict[id][0])
plt.xlabel("Date")
plt.show()
```
